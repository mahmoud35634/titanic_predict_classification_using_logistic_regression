# Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing data

```python
train_data=pd.read_csv('train.csv')
train_data
```

```python
train_data.info()
```

```python
train_data.describe()
```

# Data preprocessing

```python
train_data.isna().sum()
```

```python
sns.heatmap(train_data.isna())
```

```python
train_data['Age']=train_data['Age'].fillna(train_data['Age'].mean())
```

```python
sns.heatmap(train_data.isna())
```

```python
train_data.drop('Cabin',axis=1,inplace=True)
```

```python
train_data
```

```python
sex=pd.get_dummies(train_data['Sex'],drop_first=True)
sex
```

```python
emb=pd.get_dummies(train_data['Embarked'])
emb
```

```python
train_data=pd.concat([train_data,sex,emb],axis=1)
```

```python
train_data
```

```python
train_data.drop(labels=['PassengerId', 'Name' , 'Sex' , 'Ticket' , 'Embarked' ], axis=1 , inplace=True)
train_data
```

```python
train_data.rename(columns={'male':'Gender'},inplace=True)
train_data
```

# Explore the data

# Spliting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X=train_data.drop('Survived',axis=1)
```

```python
y=train_data['Survived']
```

```python
X
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Building the Model

<h2>LogisticRegression Algorithm</h2>

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lr=LogisticRegression(max_iter=5000)
```

```python
Lr.fit(x_train,y_train)
```

```python
Lr.score(x_train,y_train)
```

```python
y_pred=Lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

<h2> MLPClassifier Algoritmn</h2>

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(500),activation='relu',shuffle=True)
```

```python
mlp.fit(x_train,y_train)
```

```python
mlp.score(x_train,y_train)
```
